# Overview:   
The Medicare Physician & Other Practitioners - by Provider and Service dataset contains detailed information on services and procedures provided to Medicare beneficiaries by individual healthcare providers (such as physicians, nurse practitioners, physician assistants, etc.) in the United States.

This data is collected and published by the Centers for Medicare & Medicaid Services (CMS) and is used for transparency, research, and analysis of healthcare utilization and costs.

# What Does This Dataset Contain?  

Provider information: Identification details about the healthcare provider (NPI, provider name, credentials, specialty, location, etc.)  
Service information: The specific medical services or procedures performed, identified by HCPCS or CPT codes.  
Utilization measures: Counts of services rendered, number of Medicare beneficiaries served.  
Payment and charge details: Amounts billed, allowed amounts by Medicare, and payments made to the provider for each service.  
Geographical info: State, city, ZIP code of the provider.  

Data contains 9660647 rows and 28 columns

Dataset Link: https://data.cms.gov/resources/medicare-physician-other-practitioners-by-provider-and-service-data-dictionary

## Principal Component Analysis (PCA)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Fit PCA without limiting components
pca_full = PCA().fit(df_scaled)

# Compute cumulative explained variance
cumsum = np.cumsum(pca_full.explained_variance_ratio_)

# Set target cumulative variance
target_variance = 0.95

# Find the number of components needed to reach target variance
n_components = np.argmax(cumsum >= target_variance) + 1

# X-axis: component numbers starting from 1
components = np.arange(1, len(cumsum) + 1)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(components, cumsum, marker='o', linestyle='-', color='b', label='Cumulative Explained Variance')
plt.axhline(y=target_variance, color='r', linestyle='--', label=f'{int(target_variance * 100)}% Target Variance')
plt.axvline(x=n_components, color='g', linestyle='--', label=f'{n_components} Components')

# Annotate the point where target is reached
plt.annotate(f'{n_components} comps\n{cumsum[n_components - 1]:.2%} variance',
             xy=(n_components, cumsum[n_components - 1]),
             xytext=(n_components + 2, cumsum[n_components - 1] - 0.05),
             arrowprops=dict(arrowstyle="->", color='black'))

# Labels and styling
plt.xlabel('Number of PCA Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA Explained Variance vs Number of Components')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"{n_components} components are needed to reach {int(target_variance * 100)}% variance.")
```

```python
# Initialize PCA with number of components 
pca = PCA(n_components=7)

# Fit PCA on scaled data and transform
X_pca = pca.fit_transform(df_scaled)
X_pca_df = pd.DataFrame(X_pca, index=df.index)  # or df_clean.index if you cleaned data

# Explained variance ratio for each component
explained_var = pca.explained_variance_ratio_
print("Explained variance ratio by each component:", explained_var)
print(f"Total explained variance ({n_components} components):", np.sum(explained_var))
```

### Isolation Forest

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.metrics import silhouette_score, silhouette_samples
from sklearn.utils import resample
import matplotlib.pyplot as plt
```

```python
# Using PCA-transformed features (X_pca from previous step)

iso_forest = IsolationForest(
    n_estimators=100,          # Number of trees, balance speed and accuracy
    max_samples='auto',        # Use a subset for large datasets automatically
    contamination=0.01,        # Approx. 1% anomalies expected, adjust as needed
    random_state=42,
    n_jobs=-1                  # Use all processors
)

# Fit model
iso_forest.fit(X_pca)
```

```python
# Predict: -1 = anomaly, 1 = normal
df['iso_forest_label'] = iso_forest.predict(X_pca)

# Decision function scores: lower scores = more anomalous
df['iso_forest_score'] = iso_forest.decision_function(X_pca)

# Summary
print(f"Isolation Forest found {sum(df['iso_forest_label'] == -1)} anomalies.")
```

### Metrics (Isolation Forest)

```python
# Ensure X_pca is aligned with df
X_pca_df = pd.DataFrame(X_pca, index=df.index)

# Define sample size
sample_size = 100000

# Separate inliers and outliers
inliers = df[df['iso_forest_label'] == 1]
outliers = df[df['iso_forest_label'] == -1]

# Resample proportionally
sample_inliers = resample(inliers,
                          n_samples=int(sample_size * 0.9),
                          random_state=42,
                          replace=True if len(inliers) < int(sample_size * 0.9) else False)

sample_outliers = resample(outliers,
                           n_samples=int(sample_size * 0.1),
                           random_state=42,
                           replace=True if len(outliers) < int(sample_size * 0.1) else False)

# Combine
sample_df = pd.concat([sample_inliers, sample_outliers])
sample_pca = X_pca_df.loc[sample_df.index].values
sample_labels = sample_df['iso_forest_label'].values

# Compute Silhouette Score
sil_score = silhouette_score(sample_pca, sample_labels)
sample_df['silhouette'] = silhouette_samples(sample_pca, sample_labels)

# Output
print(f"\nSilhouette Score (sampled): {sil_score:.4f}")
print("\nSilhouette stats by label:")
print(sample_df.groupby('iso_forest_label')['silhouette'].describe())

print("\nDecision Function Score Summary:")
print(df['iso_forest_score'].describe())
```

```python
# Set threshold at the 1st percentile
threshold = np.percentile(df['iso_forest_score'], 1)
print(f"1st percentile decision score threshold: {threshold:.4f}")

# Count how many are considered anomalies below threshold
num_anomalies = (df['iso_forest_score'] < threshold).sum()
print(f"Number of anomalies detected (below threshold): {num_anomalies}")

# Plot histogram
plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(df['iso_forest_score'], bins=100, color='skyblue', edgecolor='black')

# Color the bins left of threshold red (anomalies)
for i in range(len(bins) - 1):
    if bins[i] < threshold:
        patches[i].set_facecolor('salmon')

# Add threshold line
plt.axvline(threshold, color='red', linestyle='--', linewidth=2, label=f'1st percentile: {threshold:.4f}')

# Optional: log scale if dominated by a few bins
# plt.yscale('log')

plt.title("Isolation Forest Decision Function Scores")
plt.xlabel("Decision Function Score")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.show()
```

### Autoencoder

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Prepare data (X_scaled) - numeric scaled features from before
X_train = df_scaled.astype('float32')

X_train_split, X_valid_split = train_test_split(X_train, test_size=0.2, random_state=42)

input_dim = X_train.shape[1]
```

```python
# Define Autoencoder architecture (simple, can be tuned)
input_layer = layers.Input(shape=(input_dim,))
encoded = layers.Dense(64, activation='relu')(input_layer)
encoded = layers.Dense(32, activation='relu')(encoded)
encoded = layers.Dense(16, activation='relu')(encoded)  # Bottleneck layer

decoded = layers.Dense(32, activation='relu')(encoded)
decoded = layers.Dense(64, activation='relu')(decoded)
decoded = layers.Dense(input_dim, activation='linear')(decoded)

autoencoder = models.Model(inputs=input_layer, outputs=decoded)

autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.summary()
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

history = autoencoder.fit(
    X_train_split, X_train_split,
    validation_data=(X_valid_split, X_valid_split),
    epochs=50,
    batch_size=1024,
    shuffle=True,
    callbacks=[early_stopping]
)
```

### Autoencoder (Metrics)

```python
# Get reconstruction errors (MSE)
X_pred = autoencoder.predict(X_train)
reconstruction_errors = np.mean(np.square(X_train - X_pred), axis=1)

# Add to dataframe
df['autoencoder_recon_error'] = reconstruction_errors

# Determine anomaly threshold (e.g., 99th percentile)
threshold = np.percentile(reconstruction_errors, 99)
print(f"Autoencoder anomaly threshold: {threshold}")

df['autoencoder_label'] = (df['autoencoder_recon_error'] > threshold).astype(int)  # 1 = anomaly

print(f"Autoencoder found {df['autoencoder_label'].sum()} anomalies.")
```

```python
# Silhouette Score for Autoencoder (on sample for speed)
from sklearn.metrics import silhouette_score, silhouette_samples

# Create pseudo-cluster labels (0 = normal, 1 = anomaly)
ae_labels = df['autoencoder_label'].values
X_sample_idx = np.random.choice(len(df_scaled), size=100000, replace=False)
X_sample = df_scaled[X_sample_idx]
ae_sample_labels = ae_labels[X_sample_idx]

# Check unique labels before computing silhouette
if len(np.unique(ae_sample_labels)) > 1:
    ae_sil_score = silhouette_score(X_sample, ae_sample_labels)
    ae_sil_samples = silhouette_samples(X_sample, ae_sample_labels)

    # Create DataFrame for analysis
    sample_df = pd.DataFrame({
        'autoencoder_label': ae_sample_labels,
        'silhouette': ae_sil_samples
    })

    # Output
    print(f"\nSilhouette Score (sampled): {ae_sil_score:.4f}")
    print("\nSilhouette stats by label:")
    print(sample_df.groupby('autoencoder_label')['silhouette'].describe())
else:
    print("Only one label present in Autoencoder. Silhouette Score cannot be calculated.")
```

```python
# Autoencoder Anomaly Percentage
ae_anomalies = (df['autoencoder_label'] == 1).sum()
ae_total = len(df)
ae_anomaly_pct = (ae_anomalies / ae_total) * 100
print(f"Autoencoder Anomalies: {ae_anomalies} ({ae_anomaly_pct:.2f}%)")
```
